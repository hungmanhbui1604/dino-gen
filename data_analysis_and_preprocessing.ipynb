```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Analysis

```python
df = pd.read_csv('data/raw/dino.csv')

print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
scientific_names = df['scientific_name']
common_names = df['common_name']
```

### Scientific Names

```python
scientific_names.head()
```

```python
unique_scientific_names = scientific_names.unique()
len(scientific_names), len(unique_scientific_names)
```

```python
scientific_word_counts = pd.Series(unique_scientific_names).astype(str).str.split().str.len()
scientific_word_counts.value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(scientific_word_counts, bins=range(min(scientific_word_counts), max(scientific_word_counts) + 2), edgecolor='black', alpha=0.7)
plt.title('Distribution of Word Counts in Unique Scientific Names')
plt.xlabel('Number of Words')
plt.xticks(range(min(scientific_word_counts), max(scientific_word_counts) + 2))
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
multi_word_scientific_names = unique_scientific_names[scientific_word_counts > 1]
multi_word_scientific_names
```

```python
single_word_scientific_names = unique_scientific_names[scientific_word_counts == 1]
scientific_character_lengths = pd.Series(single_word_scientific_names).astype(str).str.len()
scientific_character_lengths.value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(scientific_character_lengths, bins=range(min(scientific_character_lengths), max(scientific_character_lengths) + 2), edgecolor='black', alpha=0.7)
plt.title('Distribution of Character Lengths in Scientific Single Word Names')
plt.xlabel('Number of Characters')
plt.xticks(range(min(scientific_character_lengths), max(scientific_character_lengths) + 2))
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

### Common Names

```python
unique_common_names = common_names.unique()
len(common_names), len(unique_common_names)
```

```python
common_word_counts = pd.Series(unique_common_names).astype(str).str.split().str.len()
common_word_counts.value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(common_word_counts, bins=range(min(common_word_counts), max(common_word_counts) + 2), edgecolor='black', alpha=0.7)
plt.title('Distribution of Word Counts in Unique Common Names')
plt.xlabel('Number of Words')
plt.xticks(range(min(common_word_counts), max(common_word_counts) + 2))
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
multi_word_common_names = unique_common_names[common_word_counts > 1]
multi_word_common_names[:10]
```

```python
single_word_common_names = unique_common_names[common_word_counts == 1]
common_character_lengths = pd.Series(single_word_common_names).astype(str).str.len()
common_character_lengths.value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(common_character_lengths, bins=range(min(common_character_lengths), max(common_character_lengths) + 2), edgecolor='black', alpha=0.7)
plt.title('Distribution of Character Lengths in Common Single Word Names')
plt.xlabel('Number of Characters')
plt.xticks(range(min(common_character_lengths), max(common_character_lengths) + 2))
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

## Pre-processing

```python
# Step 1: Merge the two Series
merged_names = pd.concat([scientific_names, common_names], ignore_index=True)

# Step 2: Lowercase all elements
merged_names = merged_names.str.lower()

# Step 3: Keep only the first word for multi-word elements
merged_names = merged_names.str.split().str[0]

# Step 4: Remove duplicates
processed_names = merged_names.drop_duplicates()

# Step 5: Keep only names with all alphabetic characters
processed_names = processed_names[processed_names.str.isalpha()]

# Step 6: Sort the processed names and reset index
processed_names = processed_names.sort_values().reset_index(drop=True)
processed_names
```

```python
processed_names.to_csv('data/processed/dino.csv', index=False, header=False)
```

```python
char_lengths = pd.Series(processed_names).astype(str).str.len()
char_lengths.value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(char_lengths, bins=range(min(char_lengths), max(char_lengths) + 2), edgecolor='black', alpha=0.7)
plt.title('Distribution of Character Lengths in Common Single Word Names')
plt.xlabel('Number of Characters')
plt.xticks(range(min(char_lengths), max(char_lengths) + 2))
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```
